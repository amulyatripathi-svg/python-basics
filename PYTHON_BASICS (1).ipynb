```python
# Q5 - WRITE A PYTHON PROGRAM TO CHECK WHETHER A NUMBER IS EVEN OR ODD USING CONDITIONAL STATEMENTS
num = int(input("Enter a number: "))
if num % 2 == 0:
  print("Even Number")
else:
  print("Odd Number")
```

```python
# Q7 -  Create a list of five numbers and print the maximum number from the list.
numbers = [12, 45, 23, 67, 34]
maximum = max(numbers)
print("The maximum number is:", maximum)
```

```python
# Q8 - Write a python program to remove duplicate values from a list using a set.
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers))
print(unique_numbers)
```
